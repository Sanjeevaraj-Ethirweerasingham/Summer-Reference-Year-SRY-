```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/SanjeE/Desktop/SRY_1.csv')

time = np.array(df['Day_Sequence'])
temp_diff = np.array(df['Difference'])

fit = np.polyfit(time, temp_diff, 6)
regression = np.poly1d(fit)

x_range = np.linspace(1, 183, 183)

plt.scatter(time, temp_diff) # plot the data as a scatter plot
plt.plot(x_range, regression(x_range), color='red') # plot the regression line as a red line
plt.xlabel('Time') # add a label for the x-axis
plt.ylabel('Temperature Difference') # add a label for the y-axis
plt.title('6th Order Polynomial Regression') # add a title for the plot
plt.show() # display the plot
```

### Wet-bulb Temperature Adjustment


```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Define your x and y values
x = np.array([25,26,27,28,29,30])
y = np.array([23.18518519,23.79857579,24.34533898,24.7495935,25.08461538,24.26086957])

# Reshape x to a 2D array for LinearRegression.fit() method
x = x.reshape(-1, 1)

# Create a Linear Regression object
reg = LinearRegression()

# Fit the regression line to your x and y values
reg.fit(x, y)

# Get the predicted y values based on the regression line
y_pred = reg.predict(x)

# Print the slope and y-intercept of the regression line
print(f"Slope: {reg.coef_[0]:.2f}")
print(f"Y-Intercept: {reg.intercept_:.2f}")

# Plot the data points and regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('Dry-bulb Temperature')
plt.ylabel('Mean wet-bulb Temperature')
plt.show()
```
